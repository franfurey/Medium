```python
import numpy as np
import matplotlib.pyplot as plt

def plot_dotplot(data, title):
    plt.figure()
    y_vals = np.zeros_like(data)  # Set all y values to 0

    plt.scatter(data, y_vals, color="blue", marker="o", s=100, edgecolors='white')
    plt.xlabel("Value")
    plt.yticks([])
    plt.title(title)
    plt.xlim(0, 10)
    plt.show()

# Generating a smaller Gaussian dataset with a mean of 5 and a standard deviation of 1.5
np.random.seed(42)
data = np.random.normal(5, 1.5, 12)

# Adding outliers
outliers = np.array([0.5, 9.5])
data = np.concatenate((data, outliers))

# Sorting the data points for better visualization in the dot plot
data = np.sort(data)

# Plotting the dot plot
plot_dotplot(data, "Dot Plot")

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating a smaller Gaussian dataset with a mean of 5 and a standard deviation of 1.5
np.random.seed(42)
data = np.random.normal(5, 1.5, 12)

# Adding outliers
outliers = np.array([0.5, 9.5])
data = np.concatenate((data, outliers))

def plot_histogram(data, bins, title, show_data_points=False):
    plt.figure()
    counts, edges, _ = plt.hist(data, bins=bins, range=(0, 10), edgecolor="k", alpha=0.75)
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.title(title)

    if show_data_points:
        # Display the data points as circles inside the histogram bars
        bin_width = edges[1] - edges[0]
        y_vals = np.zeros_like(data)

        for i, value in enumerate(data):
            bin_index = np.where((value >= edges[:-1]) & (value <= edges[1:]))[0]
            y_vals[i] = counts[bin_index] - 0.5 * bin_width

        # plt.scatter(data, y_vals, color="red", marker="o", s=100, edgecolors='white')

    plt.xlim(0, 10)
    plt.show()

# Plotting histograms with different bin sizes
plot_histogram(data, 3, "Histogram with 3 bins", show_data_points=True)
plot_histogram(data, 6, "Histogram with 6 bins", show_data_points=True)
plot_histogram(data, 12, "Histogram with 12 bins", show_data_points=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Generating a smaller Gaussian dataset with a mean of 5 and a standard deviation of 1.5
np.random.seed(42)
data = np.random.normal(5, 1.5, 12)

# Adding outliers
outliers = np.array([0.5, 9.5])
data = np.concatenate((data, outliers))

def plot_histogram(data, bins, title, show_data_points=False):
    plt.figure()
    counts, edges, _ = plt.hist(data, bins=bins, range=(0, 10), edgecolor="k", alpha=0.75, density=True)
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.title(title)

    # Fit a Gaussian KDE to the data
    kde = gaussian_kde(data)
    x_values = np.linspace(0, 10, 1000)
    y_values = kde(x_values)

    # Plot the PDF line
    plt.plot(x_values, y_values, color="blue", linewidth=2)

    plt.xlim(0, 10)
    plt.show()

# Plotting histograms with different bin sizes
plot_histogram(data, 3, "Histogram with 3 bins")
plot_histogram(data, 6, "Histogram with 6 bins")
plot_histogram(data, 12, "Histogram with 12 bins")
```
