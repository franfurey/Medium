```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Datos de altura y peso (5 datos)
altura = np.array([150, 164, 174, 182, 190]).reshape(-1, 1)
peso = np.array([48, 70, 66, 83, 86])

# Crear el modelo de regresión lineal y entrenarlo
lr = LinearRegression()
lr.fit(altura, peso)

# Crear una regresión polinomial de grado 4 para ajustarse perfectamente a los datos
poly = PolynomialFeatures(degree=4)
altura_poly = poly.fit_transform(altura)
lr_poly = LinearRegression()
lr_poly.fit(altura_poly, peso)

# Predecir nuevos datos
altura_nueva = np.array([180, 153, 170]).reshape(-1, 1)
peso_nuevo = np.array([84, 50, 68])
peso_lr = lr.predict(altura_nueva)
peso_lr_poly = lr_poly.predict(poly.transform(altura_nueva))

# Crear el gráfico
#plt.scatter(altura, peso, color='black', label='Datos de Entrenamiento')
plt.scatter(altura_nueva, peso_nuevo, color='black', label='Datos de Testeo')
plt.plot(altura, lr.predict(altura), color='blue', linewidth=2, label='Regresión Lineal')
altura_plot = np.linspace(150, 190, 100).reshape(-1, 1)
#plt.plot(altura_plot, lr_poly.predict(poly.transform(altura_plot)), color='red', linewidth=2, label='Regresión Polinomial')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.legend()
plt.title('Comparación de Regresiones Lineales')
plt.show()
```
